```python
import numpy as np
import matplotlib.pyplot as plt
#


def Euler_method(f,t0,y0,t_final,N=0,h=None):  # inputs
    if h is None:            #initial condition
          h = (t_final - t0)/N  # step size
    else:
      N = int((t_final - t0)/h)
    t = np.zeros(N+1)
    y = np.zeros(N+1)
    t[0] = t0             # initial time
    y[0] = y0


    for i in range(N):   # loop to run the iterations
        t[i+1] = t[i] + h
        y[i+1] = y[i] + h*f(t[i],y[i]) # Euler scheme
    return t,y

def f(y,t):       # define the right hand-side of dy/dt
    return 2*y+t
def exact(t):
  return 0.25*(np.exp(2*t)-1)-0.5*t

```

```python
t,y_approx = Euler_method(f,t0=0,y0=0,t_final=1,h=0.2)
y_exact = exact(t)
abs_error = np.abs(y_exact - y_approx)
plt.plot(t,y_approx,linestyle="dashed",label='Euler Approximation')
plt.plot(t,y_exact,linestyle="solid",label='Exact Solution')
#plt.plot(t,abs_error,linestyle="dotted",label='Absolute Error')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
 #TITLE : EULER METHOD FOR PREDATOR PREY MODEL
# Define the predator-prey model (Lotka-Volterra)
def F(t, x, y, alpha, beta, delta, gamma):
  dx_dt = alpha * x - beta * x * y # Prey population equation
  dy_dt = delta * x * y - gamma * y # Predator population equation
  return dx_dt, dy_dt
# Euler method for solving the system
def euler_predator_prey(t0, tfinal, N, x0, y0, alpha, beta, delta, gamma):
  h = (tfinal - t0) / N # Step size
  t = np.linspace(t0, tfinal, N + 1) # Time grid
  x = np.zeros(N + 1) # Array to store prey population
  y = np.zeros(N + 1) # Array to store predator population
  # Initial conditions
  x[0] = x0
  y[0] = y0
  # Euler's method loop
  for k in range(N):
    dx_dt, dy_dt = F(t[k], x[k], y[k], alpha, beta, delta, gamma)
    x[k + 1] = x[k] + h * dx_dt
    y[k + 1] = y[k] + h * dy_dt
  return t, x, y
```

```python
def Heun_predator_prey(t0, t_final, N, x0, y0, alpha, beta, delta, gamma):
  t = np.linspace(t0, t_final, N + 1)
  x= np.zeros(N+1)
  y = np.zeros(N+1)

  h = (t_final-t0)/N

  #initial condition

  x[0] =x0
  y[0]=y0


  ### IMPLEMENT RK2 SCHEME
  for index in range(N):
    t[index+1] = t[index] + h
    k1x,k1y = F(t[index],x[index],y[index],alpha, beta, delta, gamma)
    k2x,k2y = F(t[index] + h,x[index]+h*k1x,y[index]+h*k1y,alpha, beta, delta, gamma)
    x[index+1] = x[index] + 0.5*h*(k1x+k2x)
    y[index+1] = y[index] + 0.5*h*(k1y+k2y)

  return t,x,y
```

```python
# Parameters for the model
alpha = 1.0 # Prey birth rate
beta = 0.1 # Predation rate
delta = 0.075 # Predator growth rate per prey eaten
gamma = 1.5 # Predator death rate
# Initial conditions and time span 1
x0 = 40 # Initial prey population
y0 = 9 # Initial predator population
t0 = 0 # Start time
tfinal = 15 # End time
N = 1000 # Number of steps
# Run the Euler solver
t, x, y = euler_predator_prey(t0, tfinal, N, x0, y0, alpha, beta, delta, gamma)
t1,x1,y1 = Heun_predator_prey(t0, tfinal, N, x0, y0, alpha, beta, delta, gamma)
# Plotting the results
# Time series plot (prey and predator over time)
plt.figure(figsize=(10, 6))
plt.plot(t, x, label="Prey Population (Euler x)", color='blue')
plt.plot(t,x1, label="Prey Population Heun (x)", color='green')
plt.plot(t, y, label="Predator Population(Euler y)", color='red')
plt.plot(t,y1, label="Predator Population Heun (y)", color='orange')
plt.xlabel("Time (t)")
plt.ylabel("Population")
plt.title("Predator-Prey Model: Time Series (Euler Method and Heun Method)")
plt.legend()
plt.grid(True)
# Phase portrait plot (predator vs prey)
plt.figure(figsize=(6, 6))
plt.plot(x, y, color='purple',label="Euler")
plt.plot(x1,y1,color="blue",label="Heun")
plt.xlabel("Prey Population (x)")
plt.ylabel("Predator Population (y)")
plt.title("Predator-Prey Model: Phase Portrait (Euler Method and Heun Method)")
plt.legend()
plt.grid(True)

plt.show()
```
